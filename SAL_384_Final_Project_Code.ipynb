```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
```

```python
def fangraphs_scrape(team, year, team_url_postfix):
    url = "https://www.fangraphs.com" + team_url_postfix + "/stats?season=" + str(year)
    print(url)
    html = urlopen(url)
    bs = BeautifulSoup(html.read(), 'lxml')
    tables = bs.find_all("table", {"class":"tablesort"})
    stats = pd.read_html(str(tables))
    
    batting = pd.DataFrame(stats[0][:-1])
    batting['Team'] = team
    batting['Season'] = year
    
    pitching = pd.DataFrame(stats[1][:-1])
    pitching['Team'] = team
    pitching['Season'] = year
    
    return(batting, pitching)
```

```python
mlb_batting_stats = pd.DataFrame()
mlb_pitching_stats = pd.DataFrame()
```

```python
for year in range(2010,2011):
    html = urlopen("https://www.fangraphs.com/teams/cubs/stats?season=" + str(year))
    bs = BeautifulSoup(html.read(), 'lxml')
    teams = bs.find("select", {"class":"select-change-team"}).find_all("option")
    for team in teams:
        team_url_postfix = team.attrs.get("value")
        if(team_url_postfix == None):
            continue
        team = team.attrs.get("label")
        batting_stats, pitching_stats = fangraphs_scrape(team, year, team_url_postfix)
        mlb_batting_stats = pd.concat([mlb_batting_stats, batting_stats])
        mlb_pitching_stats = pd.concat([mlb_pitching_stats, pitching_stats])
        time.sleep(np.random.uniform(2,5))
```

```python
mlb_batting_stats = pd.read_csv("mlb_batting_stats.csv")
mlb_pitching_stats = pd.read_csv("mlb_pitching_stats.csv")

mlb_batting_stats.PA = mlb_batting_stats.PA[mlb_batting_stats.PA > 502]
mlb_pitching_stats.IP = mlb_pitching_stats.IP[mlb_pitching_stats.IP > 162]

mlb_batting_stats = mlb_batting_stats.dropna()
mlb_pitching_stats = mlb_pitching_stats.dropna()
```

```python
mlb_batting_stats['HR_WAR_Corr'] = mlb_batting_stats['HR'].corr(mlb_batting_stats['WAR'])
mlb_batting_stats['SB_WAR_Corr'] = mlb_batting_stats['SB'].corr(mlb_batting_stats['WAR'])
mlb_batting_stats['BABIP_WAR_Corr'] = mlb_batting_stats['BABIP'].corr(mlb_batting_stats['WAR'])
mlb_batting_stats['wOBA_WAR_Corr'] = mlb_batting_stats['wOBA'].corr(mlb_batting_stats['WAR'])

mlb_batting_stats
```

```python
mlb_pitching_stats['HR_WAR_Corr'] = mlb_pitching_stats['HR/9'].corr(mlb_pitching_stats['WAR'])
mlb_pitching_stats['BABIP_WAR_Corr'] = mlb_pitching_stats['BABIP'].corr(mlb_pitching_stats['WAR'])
mlb_pitching_stats['ERA_WAR_Corr'] = mlb_pitching_stats['ERA'].corr(mlb_pitching_stats['WAR'])
mlb_pitching_stats['FIP_WAR_Corr'] = mlb_pitching_stats['FIP'].corr(mlb_pitching_stats['WAR'])

mlb_pitching_stats
```

```python
mlb_batting_war_woba_figure = plt.figure()
subplot_1 = mlb_batting_war_woba_figure.add_subplot (1,1,1)
subplot_1.set_title("wOBA vs. WAR per Season for All Hitters since 2010")
subplot_1.set_xlabel("wOBA")
subplot_1.set_ylabel("WAR")
m, b = np.polyfit(mlb_batting_stats.wOBA, mlb_batting_stats.WAR, 1)
subplot_1.plot(mlb_batting_stats.wOBA, b + m * mlb_batting_stats.wOBA, '-', color = '#D50A0A')
subplot_1.scatter(x = mlb_batting_stats.wOBA, y = mlb_batting_stats.WAR, color = '#013369', alpha = 0.1)
```

```python
mlb_pitching_war_fip_figure = plt.figure()
subplot_1 = mlb_pitching_war_fip_figure.add_subplot (1,1,1)
subplot_1.set_title("FIP vs. WAR per Season for All Pitchers since 2010")
subplot_1.set_xlabel("FIP")
subplot_1.set_ylabel("WAR")
m, b = np.polyfit(mlb_pitching_stats.FIP, mlb_pitching_stats.WAR, 1)
subplot_1.plot(mlb_pitching_stats.FIP, b + m * mlb_pitching_stats.FIP, '-', color = '#D50A0A')
subplot_1.scatter(x = mlb_pitching_stats.FIP, y = mlb_pitching_stats.WAR, color = '#013369', alpha = 0.1)
```

```python
mlb_batting_war_formula = "WAR ~ HR + SB + BABIP + wOBA"
mlb_batting_war_model = smf.ols(formula = mlb_batting_war_formula, data = mlb_batting_stats).fit()
mlb_batting_war_model.summary()
```

```python
mlb_pitching_war_formula = "WAR ~ Q('HR/9') + BABIP + ERA + FIP"
mlb_pitching_war_model = smf.ols(formula = mlb_pitching_war_formula, data = mlb_pitching_stats).fit()
mlb_pitching_war_model.summary()
```

```python
mlb_batting_war_figure, mlb_batting_war_subplot = plt.subplots()
mlb_batting_war_subplot.scatter(mlb_batting_war_model.fittedvalues, mlb_batting_war_model.resid, edgecolors = 'k', facecolors = 'none')
mlb_batting_war_subplot.set_ylabel('Residuals')
mlb_batting_war_subplot.set_xlabel('Fitted Values')
mlb_batting_war_subplot.set_title('Batting Model Residuals vs. Fitted Values')
mlb_batting_war_subplot.plot([min(mlb_batting_war_model.fittedvalues), max(mlb_batting_war_model.fittedvalues)], \
             [0,0], color = 'k', linestyle = ':', alpha = 0.3)


plt.show()
```

```python
mlb_pitching_war_figure, mlb_pitching_war_subplot = plt.subplots()
mlb_pitching_war_subplot.scatter(mlb_pitching_war_model.fittedvalues, mlb_pitching_war_model.resid, edgecolors = 'k', facecolors = 'none')
mlb_pitching_war_subplot.set_ylabel('Residuals')
mlb_pitching_war_subplot.set_xlabel('Fitted Values')
mlb_pitching_war_subplot.set_title('Pitching Model Residuals vs. Fitted Values')
mlb_pitching_war_subplot.plot([min(mlb_pitching_war_model.fittedvalues), max(mlb_pitching_war_model.fittedvalues)], \
             [0,0], color = 'k', linestyle = ':', alpha = 0.3)


plt.show()
```
